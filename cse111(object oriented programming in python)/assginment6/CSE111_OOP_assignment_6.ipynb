```python
#task1
class Student:
    ID = 0
    def __init__(self,name,dept,age,cg):
        Student.ID += 1
        self.name = name
        self.dept = dept
        self.age = age
        self.cgpa = cg
        self.id = Student.ID
    
    def showDetails(self):
        print("ID:",self.id)
        print("Name:",self.name)
        print("Department:",self.dept)
        print("Age:",self.age)
        print("CGPA:",self.cgpa)
    
    @classmethod
    def from_String(cls,mystr):
        name,dept,age,cg = mystr.split('-')
        return cls(name,dept,age,cg)
    
s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails() 
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails() 

print("A class variable is shared by all instances of the class, whereas an instance variable is unique to each instance of the class.")
print("-----------------------")
print("An instance method is bound to an instance of a class and can access and modify instance variables, while a class method is bound to the class itself and can access and modify class variables.")
```

```python
#task2
class Assassin:
    no_assassin = 0
    def __init__(self,name,success_rate):
        self.name = name
        self.success_rate = success_rate
        Assassin.no_assassin += 1
    
    def printDetails(self):
        print("Name:",self.name)
        print(f"Success rate: {self.success_rate}%")
        print("Total number of Assassin:",Assassin.no_assassin)
    
    @classmethod
    def failureRate(cls,name,fail_rate):
        success_rate = 100 - fail_rate
        return cls(name,success_rate)
    
    @classmethod
    def failurePercentage(cls,name,fail_rate_str):
        success_rate = 100 - int(fail_rate_str[:-1])
        return cls(name,success_rate)
    
john_wick = Assassin('John Wick', 100)
john_wick.printDetails()
print('================================')
nagisa = Assassin.failureRate("Nagisa", 20)
nagisa.printDetails()
print('================================')
akabane = Assassin.failurePercentage("Akabane", "10%")
akabane.printDetails()
```

```python
#task5
from datetime import datetime
class Employee:
    def __init__(self,name,exp):
        self.name = name
        self.workingPeriod = exp
    
    @classmethod
    def employeeByJoiningYear(cls,name,year):
        workingPeriod = datetime.now().year-year
        return cls(name,workingPeriod)
    
    
    @staticmethod
    def experienceCheck(exp,gender):
        if gender =='male':
            if exp>=3:
                return "He is experienced"
            else:
                return "He is not experienced"
        else:
            if exp>=3:
                return "She is experienced"
            else:
                return "She is not experienced"

employee1 = Employee('Dororo', 3)
employee2 = Employee.employeeByJoiningYear('Harry', 2016)
print(employee1.workingPeriod)
print(employee2.workingPeriod)
print(employee1.name)
print(employee2.name)
print(Employee.experienceCheck(2, "male"))
print(Employee.experienceCheck(3, "female"))
```

```python
#task6
class Laptop:
    laptopCount = 0
    def __init__(self,name,count):
        self.name = name
        self.count = count
        Laptop.laptopCount += count
    
    @classmethod
    def resetCount(cls):
        cls.laptopCount = 0
    
    @staticmethod
    def advantage():
        print("Laptops are portable")
        


lenovo = Laptop("Lenovo", 5)
dell = Laptop("Dell", 7)
print(lenovo.name, lenovo.count)
print(dell.name, dell.count)
print("Total number of Laptops", Laptop.laptopCount)
Laptop.advantage()
Laptop.resetCount()
print("Total number of Laptops", Laptop.laptopCount) 
```

```python
#task7
class Cat:
    Number_of_cats = 0
    def __init__(self,color,action):
        self.color = color
        self.action = action
        Cat.Number_of_cats += 1
        
    
    @classmethod
    def no_parameter(cls):
        return cls('White',"sitting")
    
    @classmethod
    def first_parameter(cls,color):
        return cls(color,'sitting')

    @classmethod
    def second_parameter(cls,action):
        return cls('Grey',action)
    
    def printCat(self):
        print(f"{self.color} is {self.action}")
    
    def changeColor(self,color):
        self.color = color


print("Total number of cats:", Cat.Number_of_cats)
c1 = Cat.no_parameter()
c2 = Cat.first_parameter("Black")
c3 = Cat("Brown", "jumping")
c4 = Cat("Red", "purring")
c5 = Cat.second_parameter("playing")
print("=======================")
c1.printCat()
c2.printCat()
c3.printCat()
c4.printCat()
c5.printCat()
c1.changeColor("Blue")
c3.changeColor("Purple")
c1.printCat()
c3.printCat()
print("=======================")
print("Total number of cats:", Cat.Number_of_cats)
```

```python
#task8
import math
class Cylinder:
    radius = 5
    height = 8
    def __init__(self,radius,height):
        print(f"Default radius={Cylinder.radius} and height={Cylinder.height}")
        self.radius = radius
        self.height = height
        Cylinder.radius = self.radius
        Cylinder.height = self.height
        print(f"Updated: radius={Cylinder.radius} and height={Cylinder.height}")

    @classmethod
    def changeFormat(cls,mystr):
        radius,height = mystr.split("-")
        return cls(float(radius),float(height))
    
    @classmethod
    def swap(cls,radius,height):
        return cls(height,radius)
    
    @staticmethod
    def area(radius,height):
        print(f"Area: {2*math.pi*radius**2 + 2*radius*math.pi*height}")     
    @staticmethod
    def volume(radius,height):
        print(f"Volume: {math.pi*radius**2*height}")


c1 = Cylinder(0,0)
Cylinder.area(c1.radius,c1.height)
Cylinder.volume(c1.radius,c1.height)
print("===============================")
c2 = Cylinder.swap(8,3)
c2.area(c2.radius,c2.height)
c2.volume(c2.radius,c2.height)
print("===============================")
c3 = Cylinder.changeFormat("7-13")
c3.area(c3.radius,c3.height)
c3.volume(c3.radius,c3.height)
print("===============================")
Cylinder(0.3,5.56).area(Cylinder.radius,Cylinder.height)
print("===============================")
Cylinder(3,5).volume(Cylinder.radius,Cylinder.height)
```

```python
#task10
class SultansDine:
    branch = 0
    total_sell = 0
    branch_list = []
    def __init__(self,name):
        self.name = name
        SultansDine.branch_list.append(self)
        SultansDine.branch += 1
        
        
    def sellQuantity(self,num):
        if num<10:
            self.branch_sell = num*300
        elif num<20:
            self.branch_sell = num*350
        else:
            self.branch_sell = num*400
        SultansDine.total_sell += self.branch_sell
        
    def branchInformation(self):
        print("Branch Name:",self.name)
        print("Branch Sell:",self.branch_sell)
    
    @classmethod
    def details(cls):
        print(f"Total Number of branch(s): {cls.branch}")
        print(f"Total Sell:{cls.total_sell}")
        for i in cls.branch_list:
            print(f"Branch Name: {i.name},Branch Sell: {i.branch_sell} Taka")
            print(f"Branch consists of total sell's: {((i.branch_sell/cls.total_sell)*100):.2f}%")

            
SultansDine.details()
print('########################')
dhanmondi = SultansDine('Dhanmondi')
dhanmondi.sellQuantity(25)
dhanmondi.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

baily_road = SultansDine('Baily Road')
baily_road.sellQuantity(15)
baily_road.branchInformation()
print('-----------------------------------------')
SultansDine.details()

print('========================')

gulshan = SultansDine('Gulshan')
gulshan.sellQuantity(9)
gulshan.branchInformation()
print('-----------------------------------------')
SultansDine.details()
```
