```python
#task1
class Customer:
    def __init__(self,name):
        self.name = name
    
    def greet(self,name=None):
        if name!=None:
            print(f"Hello {name}!")
        else:
            print("Hello!")
    
    def purchase(self,*items):
        print(f"{self.name}, you purchased {len(items)}")
        for i in items:
            print(i)

customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("-----------------------------")
customer_2 = Customer("David")
customer_2.greet("David")
customer_2.purchase("orange juice")
```

```python
#task7
class Match:
    def __init__(self,match):
        print("5..4..3..2..1.. Play !!!")
        match = match.split("-")
        self.batting = match[0]
        self.bowling = match[1]
        self.runs = 0
        self.wickets = 0
        self.overs = 0
    
    def add_run(self,runs):
        self.runs += runs
    
    def add_over(self,overs):
        if overs+self.overs <=5:
            self.overs += overs
        else:
            print("Warning! Cannot add 5 over/s (5 over match)")
    
    def add_wicket(self,wickets):
        self.wickets += wickets
    
    
    def print_scoreboard(self):
        return f"Batting Team: {self.batting}\nBowling Team: {self.bowling}\nTotal Runs: {self.runs} Wickets: {self.wickets} Overs: {self.overs}"

match1 = Match("Rangpur Riders-Cumilla Victorians")
print("=========================")
match1.add_run(4)
match1.add_run(6)
match1.add_over(1)
print(match1.print_scoreboard())
print("=========================")
match1.add_over(5)
print("=========================")
match1.add_wicket(1)
print(match1.print_scoreboard())
```

```python
#task9
class Batsman:
    def __init__(self,*infos):
        if len(infos)==3:
            self.name = infos[0]
            self.runs = infos[1]
            self.balls = infos[2]
        elif len(infos)==2:
            self.name = 'New Batsman'
            self.runs = infos[0]
            self.balls = infos[1]
    
    def printCareerStatistics(self):
        print("Name:",self.name)
        print("Runs Scored:",self.runs,'Balls Faced:',self.balls)
    
    def setName(self,name):
        self.name = name
    
    def battingStrikeRate(self):
        return (self.runs/self.balls)*100

b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("============================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("----------------------------")
print(b2.battingStrikeRate())
print("============================")
b1.setName("Shakib Al Hasan")
b1.printCareerStatistics()
print("----------------------------")
print(b1.battingStrikeRate())
```

```python
#task12
class TaxiLagbe:
    def __init__(self,number,location):
        self.number = number
        self.location = location
        self.passenger_list = []
        self.fare = 0
        self.passenger_num = 0
        
    def addPassenger(self,*infos):
        for i in infos:
            i = i.split('_')
            if self.passenger_num<=4:
                self.passenger_list.append(i[0])
                self.fare += int(i[1])
                self.passenger_num += 1
                print(f"Dear {i[0]}! Welcome to TaxiLagbe.")
            else:
                print("Taxi Full!No more passengers can be added.")
                
    
    def printDetails(self):
        print("Trip info for Taxi number:",self.number)
        print(f"This taxi can cover only {self.location} area.")
        print(f"Total passengers: {self.passenger_num}")
        print("Passenger lists:")
        for i in range(len(self.passenger_list)):
            if i==len(self.passenger_list)-1:
                print(self.passenger_list[i])
            else:
                print(self.passenger_list[i],end=', ')
        print("Total collected fare:",self.fare,'Taka')

taxi1 = TaxiLagbe('1010-01', 'Dhaka')
print('-------------------------------')
taxi1.addPassenger('Walker_100', 'Wood_200')
taxi1.addPassenger('Matt_100')
taxi1.addPassenger('Wilson_105')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi1.addPassenger('Karen_200')
print('-------------------------------')
taxi1.printDetails()
print('-------------------------------')
taxi2 = TaxiLagbe('1010-02', 'Khulna')
taxi2.addPassenger('Ronald_115')
taxi2.addPassenger('Parker_215')
print('-------------------------------')
taxi2.printDetails()
```

```python
#task14
class StudentDatabase:
    def __init__(self,name,id):
        self.name = name
        self.id = id
        self.grades = {}
    
    def calculateGPA(self,courses,season):
        self.courses = courses
        self.season = season
        local = []
        local_dict = {}
        sum = 0
        count = 0
        for i in courses:
            i = i.split(": ")
            local.append(i[0])
            sum += float(i[1])
            count += 1
        avg = sum/count
        local_dict[tuple(local)] = float((str(avg))[:4])
        self.grades[season] = local_dict
        
    def printDetails(self):
        for k,v in self.grades.items():
            print("Courses taken is",k+':')
            for i,j in v.items():
                for x in i:
                    print(x)
                print('GPA:',j)

s1 = StudentDatabase('Pietro', '10101222')
s1.calculateGPA(['CSE230: 4.0', 'CSE220: 4.0', 'MAT110: 4.0'], 'Summer2020')
s1.calculateGPA(['CSE250: 3.7', 'CSE330: 4.0'], 'Summer2021')
print(f'Grades for {s1.name}\n{s1.grades}')
print('------------------------------------------------------')
s1.printDetails()
s2 = StudentDatabase('Wanda', '10103332')
s2.calculateGPA(['CSE111: 3.7', 'CSE260: 3.7', 'ENG101: 4.0'], 'Summer2022')
print('------------------------------------------------------')
print(f'Grades for {s2.name}\n{s2.grades}')
print('------------------------------------------------------')
s2.printDetails()
```

```python
#task22
class Hotel:
    def __init__(self,hname):
        self.hotel_name = hname
        self.staff = 0
        self.guest = 0
        self.staff_dict = {}
        self.guest_dict = {}
    
    def addStuff(self,name,age,phone='000'):
        self.staff += 1
        print(f"Staff with ID {self.staff} is created")
        self.s_name = name
        self.s_age = age
        self.s_phone = phone
        local = {}
        local["Name"]=name
        local['Age'] = age
        local['Phone no']= phone
        self.staff_dict[self.staff] = local
    
    def addGuest(self,name,age,phone='000'):
        self.guest += 1
        print(f"Guest with ID {self.guest} is created")
        self.g_name = name
        self.g_age = age
        self.g_phone = phone
        local = {}
        local["Name"]=name
        local['Age'] = age
        local['Phone no']= phone
        self.guest_dict[self.guest] = local
    
    def getStuffById(self,id):
        return f"Staff ID: {id}\nName: {self.s_name}\nAge: {self.s_age}\nPhone no: {self.s_phone}"

    def getGuestById(self,id):
        return f"Guest ID: {id}\nName: {self.g_name}\nAge: {self.g_age}\nPhone no: {self.g_phone}"
    
    def allStaffs(self):
        print("Number of Staff:",self.staff)
        for k,v in self.staff_dict.items():
            print("Staff ID:",k,end=' ')
            for x,y in v.items():
                print(x,": ",y,end=' ')
            print()

    def allGuest(self):
        print("Number of Guest:",self.guest)
        for k,v in self.guest_dict.items():
            print("Guest ID:",k,end=' ')
            for x,y in v.items():
                print(x,": ",y,end=' ')
            print()


h = Hotel("Lakeshore")
h.addStuff( "Adam", 26)
print("=================================")
print(h.getStuffById(1))
print("=================================")
h.addGuest('Carol',35,'123')
print("=================================")
print(h.getGuestById(1))
print("=================================")
h.addGuest("Diana", 32, '431')
print("=================================")
print(h.getGuestById(2))
print("=================================")
h.allStaffs()
print("=================================")
h.allGuest()
```

```python
#task25
class Vaccine:
    def __init__(self,vac_name,location,days):
        self.vac_name = vac_name
        self.location = location
        self.days = days
    
class Person:
    def __init__(self,name,age,job='Genaral Citizen'):
        self.name = name
        self.age = age
        self.job = job
    
    def pushVaccine(self,vaccine,second=None):
        self.vaccine = vaccine.vac_name 
        if self.age<25 and self.job!="Student":
            print(f"Sorry {self.name}, Minimum age for taking vaccines if 25 years now")
        else:    
            if second==None:
                print(f"1st dose done for {self.name}")
                self.first_dose = self.vaccine
                self.second_time = f'Please come after {vaccine.days} days'
                self.first_time = 'Given'
            else:
                if self.first_dose==self.vaccine:
                    print(f"2nd dose done for {self.name}")
                    self.second_time ='Given'
                else:
                    print(f"Sorry {self.name}, you can't take 2 different vaccines")
    
    def showDetail(self):
        print("Name:",self.name,'Age:',self.age,'Type:',self.job)
        print("Vaccine name:",self.vaccine)
        print("1st dose:",self.first_time)
        print('2nd dose:',self.second_time)
 
astra = Vaccine("AstraZeneca", "UK", 60)
modr = Vaccine("Moderna", "UK", 30)
sin = Vaccine("Sinopharm", "China", 30)
p1 = Person("Bob", 21, "Student")
print("=================================")
p1.pushVaccine(astra)
print("=================================")
p1.showDetail()
print("=================================")
p1.pushVaccine(sin, "2nd Dose")
print("=================================")
p1.pushVaccine(astra, "2nd Dose")
print("=================================")
p1.showDetail()
print("=================================")
p2 = Person("Carol", 23, "Actor")
print("=================================")
p2.pushVaccine(sin)
print("=================================")
p3 = Person("David", 34)
print("=================================")
p3.pushVaccine(modr)
print("=================================")
p3.showDetail()
print("=================================")
p3.pushVaccine(modr, "2nd Dose")
```
