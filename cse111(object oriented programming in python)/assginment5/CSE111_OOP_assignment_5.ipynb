```python
#task1
import math
class Circle:
    def __init__(self,radius=None):
        self.setRadius(radius)
    
    def setRadius(self,radius):
        self.__radius = radius
        
    def getRadius(self):
        return self.__radius
    
    def area(self):
        return math.pi*self.getRadius()**2


    
c1 = Circle(4)
print("First circle radius:" , c1.getRadius())
print("First circle area:" , c1.area())


c2 = Circle(5)
print("Second circle radius:" , c2.getRadius())
print("Second circle area:" , c2.area())
```

```python
#task2
class Triangle:
    def __init__(self,base=None,height=None):
        self.setBase(base)
        self.setHeight(height)
    
    def setBase(self,base):
        self.__base = base
    
    def setHeight(self,height):
        self.__height = height
    
    def getBase(self):
        return self.__base
    
    def getHeight(self):
        return self.__height
    
    def area(self):
        return 0.5*self.getBase()*self.getHeight()


t1 = Triangle(10, 5)
print("First Triangle Base:" , t1.getBase())
print("First Triangle Height:" , t1.getHeight())
print("First Triangle area:" ,t1.area())


t2 = Triangle(5, 3)
print("Second Triangle Base:" , t2.getBase())
print("Second Triangle Height:" , t2.getHeight())
print("Second Triangle area:" ,t2.area())
```

```python
#task5
class Student:
    def __init__(self,name,id,dept):
        self.setName(name)
        self.setId(id)
        self.setDept(dept)
        print("Student object is created!")
        self.email = None
        self.password = None
        self.login = False
        
        
    def setName(self,name):
        self.__name = name
    
    def setId(self,id):
        self.__id = id
    
    def setDept(self,dept):
        self.__department = dept
    
    def getName(self):
        return self.__name
    
    def getId(self):
        return self.__id
    
    def getDept(self):
        return self.__department

class Usis:
    def __init__(self):
        print("USIS is ready to use!")
    
    def login(self,user):
        if user.email ==None and user.password==None:
            print("Email and password need to be set.")
        else:
            print("Login successful!")
            user.login = True
    
    def advising(self,user,*courses):
        if user.login:
            if len(courses)==0:
                print("You haven't selected any courses.")
            elif len(courses)==3:
                print("Advising successful!")
            elif len(courses)>3:
                print("You need special approval to take more than 3 courses.")
            self.courses = ''
            for i in range(len(courses)):
                if i== len(courses)-1:
                    self.courses += courses[i]
                else:
                    self.courses += courses[i] + ', '
        else:
            print("Please login to advise courses!")

    def individualDetails(self,user):
        return f"Name: {user.getName()}\nID: {user.getId()}\nDepartment: {user.getDept()}\nAdvised courses: {self.courses}"
        
            
            
    


rakib = Student("Rakib", 12301455, "CSE")
print("1***********************")
usis_obj = Usis()
print("2***********************")
usis_obj.login(rakib)
print("3***********************")
usis_obj.advising(rakib)
print("4***********************")
rakib.email = "rakib@hotmail.com"
rakib.password = "1234"
print("5***********************")
usis_obj.login(rakib)
print("6***********************")
usis_obj.advising(rakib)
print("7***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110", "CSE260")
print("8***********************")
usis_obj.advising(rakib, "CSE110", "PHY111", "MAT110")
print("9***********************")
print(usis_obj.individualDetails(rakib))
```

```python
#task6
class Train:
    def __init__(self,name,*locations):
        self.locations = locations
        self.start = locations[0]
        self.end = locations[-1]
        print(f"Welcome aboard on {name}")
        print("Start:",self.start)
        print("Destination:",self.end)
        self.passengers = []
    
    def addPassenger(self,*passengers):
        for i in passengers:
            local = {}
            print(f"{i.name} welcome aboard")
            if len(i.infos)==0:
                i.starting = self.start
                i.ending = self.end
            elif len(i.infos)==1:
                i.starting = i.infos[0]
                i.ending = self.end
            elif len(i.infos)==2:
                i.starting = i.infos[0]
                i.ending = i.infos[1]
            local['Name'] = i.name
            local['Start'] = i.starting
            local['Destination'] = i.ending
            index1 = 0
            index2 = 0
            for x in range(len(self.locations)):
                if i.starting == self.locations[x]:
                    index1 = x
                if i.ending == self.locations[x]:
                    index2 = x
            local['Fair'] = 100*(index2-index1)
            self.passengers.append(local)
        
    def allPassengerDetails(self):
        for i in self.passengers:
            for k,v in i.items():
                if k=='Fair':
                    print(f"{k}: ${v}")
                else:
                    print(f"{k}: {v}",end=', ')
        
            
                
            
                
                
    
class Passenger:
    def __init__(self,name,*infos):
        self.infos = infos
        self.name = name
        

t1 = Train('T1-Express','New York','Manhattan','Brooklyn','Boston')
print("1========================")
p1 =Passenger("Naruto")
t1.addPassenger(p1)
p2 = Passenger("Sasuke","Manhattan")
p3 = Passenger("Hinata","Manhattan","Brooklyn")
print("2========================")
t1.addPassenger(p2,p3)
print("3========================")
t1.allPassengerDetails()
print("4========================")
t2 = Train('Europe-Express','London','Paris','Brussels','Turkey')
print("5========================")
p4 =Passenger("Max","London","Brussels")
p5 = Passenger("Eleven","Paris")
p6 = Passenger("Mike","Brussels")
t2.addPassenger(p4,p5,p6)
print("6========================")
t2.allPassengerDetails()
```

```python
#task7
class BracuStudent:
    def __init__(self,name,home):
        self.name = name
        self.home = home
        self.passinfo = False


    def get_pass(self):
        self.passinfo = True


        
    def show_details(self):
        print("Name:",self.name)
        print("Location:",self.home)
        print("have Bus Pass?",self.passinfo)

class BracuBus:
    def __init__(self,route,max=2):
        self.max = max
        self.route = route
        self.passengers = []
    
    
    def show_details(self):
        print("Bus Route:",self.route)
        print(f"Passengers Count: {len(self.passengers)} (Max: {self.max})")
        print("Passengers On Board:",self.passengers)
        
    
    def board(self,*passengers):
        if len(passengers)==0:
            print("No passenger!")
        else:
            for i in passengers:
                if i.passinfo:
                    if i.home==self.route:
                        if len(self.passengers)<self.max:
                            if i.name not in self.passengers:
                                self.passengers.append(i.name)
                                print(f"{i.name} boarded the bus.")
                        else:
                            print("Bus is full!") 
                    else:
                        print("You got on wrong bus!")
                else:
                    print("You don't have bus pass!")
                    
        
        

st1 = BracuStudent("Afif", "Mirpur")
print("1===========================")
st2 = BracuStudent("Shanto", "Motijheel")
st3 = BracuStudent("Taskin", "Mirpur")
st1.show_details()
st2.show_details()
print("2===========================")
st3.show_details()
print("3===========================")
bus1 = BracuBus("Mirpur")
bus2 = BracuBus("Azimpur", 5)
bus1.show_details()
bus2.show_details()
print("4===========================")
st2.get_pass()
st3.get_pass()
print("5===========================")
st2.show_details()
st3.show_details()
print("6===========================")
bus1.board()
print("7===========================")
bus1.board(st1, st2)
print("8===========================")
st1.get_pass()
st2.home = "Mirpur"
st1.show_details()
st2.show_details()
print("9===========================")
bus1.board(st1, st2, st3)
print("10===========================")
bus1.show_details()
```

```python
#task8
class Library:
    def __init__(self,location,book_dict):
        self.location = location
        self.book_dict = book_dict
        self.borrower = {}
    
    def details(self):
        print(f"{self.location} Library details")
        print("Borrower details:")
        print(self.borrower)
        print("Books availability:")
        print(self.book_dict)
    
class Reader:
    def __init__(self,name):
        self.name = name
        self.total_books = 0
        self.person_info = {}
    
    def borrow(self,library,*books):
        for i in books:
            if self.total_books<5:
                if library.book_dict[i]<=0:
                    print(f"{i} books are not available at the moment.")
                else:
                    print(f"{i} book is borrowed successfully.")
                    self.total_books += 1
                    if i not in self.person_info.keys():
                        self.person_info[i] = 1
                    else:
                        self.person_info[i] += 1
                    library.book_dict[i] -=1
            else:
                print("You cannot borrow more than 5 books.")
                
        library.borrower[self.name] = self.total_books
    
    def readerInfo(self,info=None):
        if info==None:
            print(f"{self.name}, you have {self.total_books} book(s) with you.")
            for k,v in self.person_info.items():
                print(f"Books on {k}: {v}")
        else:
            print(f"{self.name}, you have {self.person_info[info]} {info} book(s) with you")
            

L1=Library('Dhaka',{'Arts':15,'Fiction':135,'Politics':2,'Science':11,'Poetry':15})
L1.details()
print("1----------------------")
r1=Reader('Aladdin')
r1.borrow(L1,'Arts','Fiction','Fiction','Politics')
print("2----------------------")
r1.borrow(L1,'Politics','Fiction')
print("3----------------------")
r1.readerInfo()
print("4----------------------")
r1.readerInfo('Fiction')
print("5----------------------")
L1.details()
print("6----------------------")
r2=Reader('Jasmine')
r2.borrow(L1,'Politics','Poetry')
print("7----------------------")
r2.readerInfo()
print("8----------------------")
L1.details()
```
