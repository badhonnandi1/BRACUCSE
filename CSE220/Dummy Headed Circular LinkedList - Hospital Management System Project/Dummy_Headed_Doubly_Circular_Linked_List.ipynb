```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:

  # Creates a Dummy Head for Dummy Headed Circular Doubly Circular Linked List.
  def __init__(self):
    # To Do
    self.head = Node(None,None,None)

  #create a DHDCL from an array
  def create_list(self, l):
    # To Do
    for i in range(len(l)):
      node = Node(l[i],None,None)

      if self.head.next == None:
        self.head.next = node
        node.prev = self.head
        self.head.prev = node
        node.next = self.head
      else:
        prev1 = self.head.prev
        prev1.next = node
        node.prev = prev1
        self.head.prev = node
        node.next = self.head



  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    # To Do
    count = 0
    temp = self.head.next
    while temp != self.head:
      count += 1
      temp = temp.next

    return count


  # prints the elements in the list forward
  def forwardprint(self):
    # To Do
    mystr = ''
    temp = self.head.next
    while temp != self.head:
      mystr += str(temp.element)
      if temp.next != self.head:
        mystr += ','
      else:
        mystr += '.'
      temp = temp.next
    print(mystr)


  # prints the elements in the list backward
  def backwardprint(self):
    # To Do
    mystr = ''
    temp = self.head.prev
    while temp!=self.head:
      mystr += str(temp.element)
      if temp.prev != self.head:
        mystr += ','
      else:
        mystr += '.'
      temp = temp.prev
    print(mystr)

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    # To Do
    if idx<0:
      return None
    count = 0
    temp = self.head.next
    for i in range(idx):
      temp = temp.next
    return temp

  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    # To Do
    count = 0
    flag = False
    temp = self.head.next
    while temp!=self.head:
      if temp.element == elem:
        flag = True
        break
      count +=1
      temp = temp.next

    if flag:
      return count
    else:
      return None

  # insert at specific index of the DHDCL
  def insert(self, elem, idx):
    # To Do
    newNode = Node(elem,None,None)
    if idx ==0:
      nextNode = self.head.next
      newNode.next = nextNode
      nextNode.prev = newNode
      self.head.next = newNode
      newNode.prev = self.head
    else:
      prev = self.head.next
      for i in range(idx-1):
        prev = prev.next
      nextNode = prev.next
      newNode.next = nextNode
      newNode.prev = prev
      prev.next = newNode
      nextNode.prev = newNode



  #removes the node of the specific index and returns the element of the node
  def remove(self, idx):
    # To Do
    popped_node = self.head.next
    for i in range(idx):
      popped_node = popped_node.next
    prevNode = popped_node.prev
    nextNode = popped_node.next
    prevNode.next = nextNode
    nextNode.prev = prevNode
    popped_node.prev = None
    popped_node.next = None

    return popped_node.element
```
