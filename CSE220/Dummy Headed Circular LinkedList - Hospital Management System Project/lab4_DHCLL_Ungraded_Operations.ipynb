```python
class Patient:
    def __init__(self, id, name, age, blood):
        self.id = id
        self.name = name
        self.age = age
        self.blood = blood

class Node:
    def __init__(self, value=None):
        self.value = value
        self.prev = None
        self.next = None

class WRM:
    def __init__(self):
        self.head = Node()
        self.patientnumber = 0

    def RegisterPatient(self):
        id = int(input("Patient ID: "))
        name = input("Patient Name: ")
        age = int(input("Patient Age: "))
        blood = input("Patient blood group: ")

        patient_obj = Patient(id, name, age, blood)
        patientNode = Node(patient_obj)


        if self.patientnumber == 0:
            self.head.next = patientNode
            self.head.prev = patientNode
            patientNode.prev = self.head
            patientNode.next = self.head
        else:
            lastPatient = self.head.prev
            patientNode.prev = lastPatient
            patientNode.next = self.head
            lastPatient.next = patientNode
            self.head.prev = patientNode
        self.patientnumber += 1
        print("Paitent successfully admitted")

    def ShowAllPatient(self):
        temp = self.head.next
        mystr = ""
        while temp != self.head:
            mystr += temp.value.name
            if temp.next != self.head:
                mystr += "-->"
            temp = temp.next
        return mystr

    def ServePatient(self):
        if self.patientnumber == 0:
            print("No patients left")
        else:
            patient = self.head.next
            if self.patientnumber == 1:
                self.head.prev = self.head
                self.head.next = self.head
            else:
                self.head.next = patient.next
                patient.next.prev = self.head
            patient.next = None
            patient.prev = None
            self.patientnumber -= 1
            print(f"{patient.value.name} surved succesfully")

    def CancelAll(self):
        self.head.next = self.head
        self.head.prev = self.head
        self.patientnumber = 0
        print("All the appointments got cancelled succesfully")

    def CanDoctorGoHome(self):
        if self.patientnumber == 0:
            print("Yes the doctor is free to go")
        else:
            print("There are many patients left in the hospital he is not free")

    def ReverseTheLine(self):
        current = self.head
        while True:
            temp = current.next
            current.next = current.prev
            current.prev = temp
            current = temp
            if current == self.head:
                break

        print("Successfully reversed the patients queue!!")


hospital = WRM()
while True:

    print(
        """
        Hello People , welcome to hospital management system app.
        Please Enter number as the instraction given below.
        Press 1 for Resistering new Patient
        Press 2 to Server a Patient
        Press 3 to show the Patient queue
        Press 4 for checking if the doctor is free or not?
        Press 5 to cancel all the appointments
        Press 6 to reverse the queue of patient
        Press 7 to exit the system
"""
    )
    user = int(input())
    if user == 1:
        hospital.RegisterPatient()

    elif user == 2:
        hospital.ServePatient()

    elif user == 3:
        print(hospital.ShowAllPatient())

    elif user == 4:
        hospital.CanDoctorGoHome()

    elif user == 5:
        hospital.CancelAll()

    elif user == 6:
        hospital.ReverseTheLine()

    elif user == 7:
      print("Thanks for using the hospital management app!!")
      break

    else:
        print("Please Enter a valid input!!!")
```
